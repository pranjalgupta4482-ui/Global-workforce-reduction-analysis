```python
import matplotlib.pyplot as plt
df.groupby("year")["number_laid_off"].sum().plot(kind="bar")
plt.title("Total workforce Reduction by year")
plt.xlabel("year")
plt.ylabel("number_lay_offs")
plt.show()
```

```python
import seaborn as sns

top_industries = df.groupby("industry")["percentage_of_workforce"].mean().sort_values(ascending=False).head(10)

sns.barplot(x=top_industries.values, y=top_industries.index)

plt.title("Top Industries by Workforce Impact (%)")
plt.xlabel("Impact %")
plt.ylabel("Industry")
plt.show()
```

```python
top_companies = df.groupby("company")["number_laid_off"].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_companies.values, y=top_companies.index)

plt.title("Top Companies by Workforce Reduction")
plt.xlabel("Total Laid Off")
plt.ylabel("Company")
plt.show()
```

```python
top_industries = df["industry"].value_counts().head(10).index

sns.boxplot(
    x="number_laid_off",
   y="industry",
    data=df[df["industry"].isin(top_industries)]
)
```
